```python
import numpy as np
data = np.array([
    [53,500],
    [22,45],
    [5,50],
    [3,0],
    [7,160],
    [1,47],
    [9,580],
    [0,600],
    [0,155],
    [5,50]
])
age= np.array([53,22,5,3,7,1,9,0,0,5])
income = np.array([500,45,50,0,160,47,580,600,155,50])
print(age)
```

```python
def Euclidean(a,b):
    return np.sqrt(np.sum((a-b)**2))
def Manhattan(a,b):
    return np.sum(np.abs(a-b))
def Minkowski(a,b,p=3):
    return np.power(np.sum(np.abs(a-b)**p),1/p)
def Chebyshev(a,b):
    return np.max(np.abs(a-b))
```

```python
def dissimilarity_matrix(data,dis_fun,p=None):
    n = data.shape[0]
    matrix = np.zeros((n,n))
    for i in range (n):
        for j in range (i,n) :
            if dis_fun == Minkowski:
                dissimilarity = Minkowski(data[i],data[j],p=3)
            else:
                dissimilarity = dis_fun(data[i],data[j])
            matrix[j,i]=dissimilarity
    return matrix
```

```python
#FOR AGE AND INCOME
Euclidean_matrix = dissimilarity_matrix(data,Euclidean)
Manhattan_matrix = dissimilarity_matrix(data,Manhattan)
Minkowski_matrix = dissimilarity_matrix(data,Minkowski,p=3)
Chebyshev_matrix = dissimilarity_matrix(data,Chebyshev)
```

```python
print("Euclidean \n",Euclidean_matrix)
print("Manhattan \n",Manhattan_matrix)
print("Minkowski \n",Minkowski_matrix)
print("Chebyshev \n",Chebyshev_matrix)
```

```python
#FOR AGE
Euclidean_matrix_age = dissimilarity_matrix(age,Euclidean)
Manhattan_matrix_age = dissimilarity_matrix(age,Manhattan)
Minkowski_matrix_age = dissimilarity_matrix(age,Minkowski,p=3)
Chebyshev_matrix_age = dissimilarity_matrix(age,Chebyshev)
```

```python
print("Euclidean \n",Euclidean_matrix_age)
print("Manhattan \n",Manhattan_matrix_age)
print("Minkowski \n",Minkowski_matrix_age)
print("Chebyshev \n",Chebyshev_matrix_age)
```

```python
#FOR INCOME
Euclidean_matrix_income = dissimilarity_matrix(income,Euclidean)
Manhattan_matrix_income = dissimilarity_matrix(income,Manhattan)
Minkowski_matrix_income = dissimilarity_matrix(income,Minkowski,p=3)
Chebyshev_matrix_income = dissimilarity_matrix(income,Chebyshev)
```

```python
print("Euclidean \n",Euclidean_matrix_income)
print("Manhattan \n",Manhattan_matrix_income)
print("Minkowski \n",Minkowski_matrix_income)
print("Chebyshev \n",Chebyshev_matrix_income)
```

```python
#WEIGHTED 
def weighted_dissimilarity_matrix(data,dis_fun,weight=1,p=None):
    n = data.shape[0]
    matrix = np.zeros((n,n))
    for i in range (n):
        for j in range (i,n) :
            if dis_fun == Minkowski:
                dissimilarity = Minkowski(weight*data[i],data[j],p=3)
            else:
                dissimilarity = dis_fun(weight*data[i],data[j])
            matrix[j,i]=dissimilarity
    return matrix
```

```python
Euclidean_matrix_w = weighted_dissimilarity_matrix(data,Euclidean,2.5)
Manhattan_matrix_w = weighted_dissimilarity_matrix(data,Manhattan,2.5)
Minkowski_matrix_w = weighted_dissimilarity_matrix(data,Minkowski,weight=2.5,p=3)
Chebyshev_matrix_w = weighted_dissimilarity_matrix(data,Chebyshev,2.5)
print("Euclidean \n",Euclidean_matrix_w)
print("Manhattan \n",Manhattan_matrix_w)
print("Minkowski \n",Minkowski_matrix_w)
print("Chebyshev \n",Chebyshev_matrix_w)
```
